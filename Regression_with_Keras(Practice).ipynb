```python
import numpy as np
import matplotlib.pyplot as plt
```

# y = a*x + b + noise

```python
a = 5 ; b = 2

x = np.linspace(0,50,500)

noise = np.random.normal(loc=0,scale=10,size=len(x))

y = a*x + b + noise
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,'*')
plt.grid()
plt.xlabel('X') ; plt.ylabel('Y')
```

# ANN-Implementation

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

### 1 ---> 4 ----> 4 ----> 1

```python
#Create the ANN now. 

#1. 1--->4
model.add(Dense(4,input_dim=1,activation='relu'))

#2. 4--->4
model.add(Dense(4,activation='relu'))

#3. 4---->1
model.add(Dense(1,activation='linear'))

model.compile(loss='mse',optimizer='adam')
```

```python
model.summary()
```

### Fitting

```python
model.fit(x,y,epochs=200)
```

```python
loss = model.history.history['loss']
epochs = list(range(len(loss)))

plt.plot(epochs,loss)
plt.ylabel('Loss'); plt.xlabel('# Epochs') ; plt.grid()
```

```python
yp = model.predict(x)
```

## Visual and Statistical Evaluation of the Model

```python
plt.style.use('default')
plt.figure(figsize=(8,6))
plt.plot(x,y,'*',label='Actual Data',alpha=0.3,color='green')

plt.plot(x,yp,'red',label='Neural-Network fit Model',lw=3) 

plt.grid()

plt.xlabel('Height of a specie(cms)')
plt.ylabel('Weight(lbs)')
plt.legend()
```
