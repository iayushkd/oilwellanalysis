```
import matplotlib.pyplot as plt
%matplotlib inline
```

#Interpolation

```
import numpy as np
from scipy import interpolate,integrate
```

```
x = np.arange(-1,11)
```

```
y = np.exp(-x/3.0)
```

```
plt.style.use('fivethirtyeight')
plt.plot(x,y)
```

```
#Step 1: Find the Interpolated Function:
```

```
f = interpolate.interp1d(x,y)
```

```
#values from interpolated function
l = []
for i in x:
  v = float(f(i))
  l.append(v)

y_int = np.array(l)
```

```
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(x,y_int,'r')
plt.title('Interpolated function')
plt.subplot(1,2,2)
plt.plot(x,y,'b')
plt.title('Actual Function')
plt.tight_layout(False)
```

#Comparing the Actual function and Inerpolated function using integration.
We do this since we cannot see any visible difference. Hence The integrated values are the best way to compare

```
ans,err = integrate.quad(f,0,10)
print(f"1.The integration of interpolated fn gives {ans}")

intg = integrate.simps(y[1:-1],x[1:-1])
print(f"1.The integration of actual fn gives {intg}")
```

```
#Conclusion : Interpolation can be said to be a pretty good estimate of the actual function!
```
