#Gradient Descent to find the minima of a multivariable function z = f(x,y)

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```
x = np.arange(-100,101)
y = np.arange(-100,101)
z = x**2 + y**2
```

```
plt.figure(figsize=(8,6))
ax1 = plt.subplot(111, projection='3d') 
ax1.plot(x,y,z)
plt.xlabel('X values')
plt.ylabel('Y values')
```

```
alpha = 0.1

x_r = 50
y_r = 50

z_r = [0,5000]

#delz_delx = 2x
#delz_dely = 2y

while abs(z_r[-1] - z_r[-2]) >= 0.000000000001:
  x_r = x_r - alpha*2*x_r
  y_r = y_r - alpha*2*y_r
  zr = x_r**2 + y_r**2
  z_r.append(zr)




```

```
print(f'1. The value of x that gives a minima is: {x_r}')
print(f'2. The value of y that gives a minima is: {y_r}')
print(f'3. The minimum value of of the function is: {z_r[-1]}')
```

```
#Observe how close these values are to the real solutions. 
#Real solutions- 1. x_r = 0 , 2. y_r = 0, 3. z_min = 0 (Very close convergence)
```
